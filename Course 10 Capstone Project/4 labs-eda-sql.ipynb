## Tasks

Now write and execute SQL queries to solve the assignment tasks.

### Task 1

##### Display the names of the unique launch sites  in the space mission


```python
df.loc[:,"Launch_Site"].value_counts().index
```

### Task 2

##### Display 5 records where launch sites begin with the string 'CCA'


```python
import re
```

```python
def find_CCA(site):
	if re.search(pattern=r'^CCA', string=site) != None:
		return True
	return False

df.loc[df.loc[:,"Launch_Site"].apply(find_CCA)].head()
```

### Task 3

##### Display the total payload mass carried by boosters launched by NASA (CRS)


```python
def from_NASA(client):
	if re.search(pattern=r'^NASA', string=client) != None:
		return True
	return False

df.loc[df.loc[:,"Customer"].apply(from_NASA),"PAYLOAD_MASS__KG_"].sum()
```

### Task 4

##### Display average payload mass carried by booster version F9 v1.1


```python
def boosterModel(booster):
	if re.search(pattern=r'^F9 v1.1', string=booster) != None:
		return True
	return False

df.loc[df.loc[:,"Booster_Version"].apply(boosterModel),"PAYLOAD_MASS__KG_"].mean()
```

### Task 5

##### List the date when the first successful landing outcome in ground pad was acheived.

*Hint:Use min function*


```python
df.Date = pd.to_datetime(arg=df.Date, dayfirst=True)
```

```python
def landingSuccess(landing):
	if re.search(pattern=r'Success', string=landing) != None:
		return True
	return False

df.loc[df.loc[:,"Landing _Outcome"].apply(landingSuccess),:].sort_values(by="Date", axis=0).loc[:,"Date"].head(1)
```

### Task 6

##### List the names of the boosters which have success in drone ship and have payload mass greater than 4000 but less than 6000


```python
df_4000_6000 = df[df.loc[:,"PAYLOAD_MASS__KG_"].between(left=4000, right=6000)]
result = df_4000_6000[df_4000_6000.loc[:,"Mission_Outcome"].str.contains(pat=r'Success')].loc[:,"Booster_Version"].values
print(result)
```

### Task 7

##### List the total number of successful and failure mission outcomes


```python
df.loc[:,"Mission_Outcome"].str.contains(pat=r"Success").value_counts()
```

### Task 8

##### List the   names of the booster_versions which have carried the maximum payload mass. Use a subquery


```python
maxLoad = df.loc[:,"PAYLOAD_MASS__KG_"].max()
df.loc[df.loc[:,"PAYLOAD_MASS__KG_"] == maxLoad, :].loc[:,"Booster_Version"]
```

### Task 9

##### List the failed landing_outcomes in drone ship, their booster versions, and launch site names for in year 2015


```python
droneShip = df.loc[:,"Landing _Outcome"].str.contains(pat=r"drone")
after2015 = df.loc[:,"Date"].dt.year >= 2015
df.loc[droneShip & after2015, ["Landing _Outcome", "Booster_Version", "Launch_Site"]]
```

### Task 10

##### Rank the count of landing outcomes (such as Failure (drone ship) or Success (ground pad)) between the date 2010-06-04 and 2017-03-20, in descending order


```python
# df.loc[:,"Date"].dt.year.between().loc[:,"Landing _Outcome"].value_counts(ascending=False)
condition_1 = df.loc[:,"Date"] > "2010-06-04" 
condition_2 = df.loc[:,"Date"] < "2017-03-20"
df.loc[condition_1 & condition_2].loc[:,"Landing _Outcome"].value_counts(ascending=False)
```
